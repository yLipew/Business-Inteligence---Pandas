```python
#IMPORT's
import pandas as pd
import numpy as np
```

```python
a = pd.Series()
print(a)
```

```python
#Não passamos nenhum índice, então, por padrão, ele atribuiu os índices que variam de 0 a len(data) − 1 , ou seja, 0 a 3.
b = pd.Series([10,50,70,40,57,60,77])
print(b)
```

```python
#Dessa vez passamos os valores do índice.
c = np.array(["a","b","c","d"])
resultado = pd.Series(c)
print(resultado)
```

```python
d = np.array(["a","b","c","d"])
resultado = pd.Series(d, index=[10,20,30,40])
print(resultado)
```

```python
#As chaves de dicionário são usadas
para construir o índice.
e = {"a": 10., "b": 20., "c": 30., "d": 40.}
resultado = pd.Series(e)
print(resultado)
```

```python
#Observe que a ordem do índice é mantida e o elemento ausente é preenchido com NaN (Not a Number).

f = {"a": 100., "b": 200., "c": 300.}
resultado = pd.Series(f, index=["c", "b", "a", "d"])
print(resultado)
```

```python
#Se os dados forem um valor escalar, um índice deverá ser fornecido. O valor será repetido para corresponder ao comprimento do índice.
g = pd.Series(7, index=[1,2,3,4,5])
print(g)
```

```python
#Vamos buscar o primeiro elemento.
h = pd.Series([10,20,30,40,50],index=["a", "b", "c", "d", "e"])
print(h)


#Buscando o primeiro elemento ou qualquer que queira consultar
#ACESSO POR POSIÇÃO (Recomendado para loops ou ordem fixa)
print ("O elemento pesquisado é: ",h.iloc[0])

#PULANDO LINHA
print()

# ACESSO POR RÓTULO (Recomendado para clareza e filtros específicos)
print("O elemento pesquisado é: ",h.loc["e"]) # Retorna 50
```

```python
i = pd.Series([1,2,3,4,5],index=["a", "b", "c", "d", "e"])

#Buscando os três primeiros elementos
print(i[:3])

#PULANDO LINHA
print()

#Buscando os três últimos elementos
print(i[-3:])
```

```python
#Recuperar dados usando rótulo (Index)
j = pd.Series([1,2,3,4,5],index=["a", "b", "c", "d", "e"])

print(j["d"])
```

```python
#Imprimir vários elementos usando uma lista de valores de rótulo de índice.

k = pd.Series([1,2,3,4,5],index = ["a", "b", "c", "d", "e"])

print(k[["c","a","e"]])
```

```python
#Se um rótulo não existir, uma exceção será gerada. (KeyError)

l = pd.Series([1,2,3,4,5],index = ["a", "b", "c", "d", "e"])

print(l[["m"]])
```
